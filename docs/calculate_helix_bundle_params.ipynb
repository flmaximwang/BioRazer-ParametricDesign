```python
from biorazer.structure.io.protein import PDB2STRUCT
from br_parametric_design.models.helix import CCCPHelixBundle

atom_array = PDB2STRUCT(input_file="1a3n.pdb").read()
helix_bundle = CCCPHelixBundle.from_mask(atom_array, mask={
    "helix_1": (atom_array.res_id >= 1) & (atom_array.res_id <= 10),
    "helix_2": (atom_array.res_id >= 11) & (atom_array.res_id <= 20),
    "helix_3": (atom_array.res_id >= 21) & (atom_array.res_id <= 30),
    "helix_4": (atom_array.res_id >= 31) & (atom_array.res_id <= 40),
})
helix_bundle.fit()
```
